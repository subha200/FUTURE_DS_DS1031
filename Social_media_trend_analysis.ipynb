##Removing unnecessary columns and coverting the time-stamp format

```python
# Droping unnecessary index columns
df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"], inplace=True)

# Converting Time-stamp to datetime format
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

# Checking the cleaned dataset
df.info(), df.head()
```

##Standardizing the platform and country names them for better analysis

```python
print("Before Cleaning:", df["Platform"].unique())  # View original platform names

# Standardize Platform names
df["Platform"] = df["Platform"].str.strip().str.capitalize()

print("After Cleaning:", df["Platform"].unique())  # View cleaned platform names

df["Country"] = df["Country"].str.strip().str.capitalize()
print("Country names after Cleaning:", df["Country"].unique())
```

##**Step 3: Cleaning Text Column (For NLP & Sentiment Analysis)**



###The Text column might contain:

1.   URLs (links that are not useful for sentiment analysis)
2.   Special characters & punctuation
3. Emojis & extra spaces
4. Convert text to lowercase (for consistency)







```python
import re

def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)  # Remove URLs
    text = re.sub(r"[^a-zA-Z0-9\s#@]", "", text)  # Remove special characters except hashtags
    text = text.strip()  # Remove extra spaces
    return text

df["Text"] = df["Text"].apply(clean_text)

df['Text']
```

###Dividing unique sentiments into categories for better standardization for proper analysis

```python
# Define sentiment categories for standardization with more sentiments
very_positive_labels = ["Joy", "Excitement", "Adrenaline", "Love", "Happy", "Great",
                        "Amazing", "Excellent", "Wonderful", "Fantastic", "Beautiful",
                        "Cheerful", "Excited", "Optimistic", "Enthusiastic", "Thrilled",
                        "Delighted", "Inspired", "Grateful"]  # Strong positive sentiments

positive_labels = ["Positive", "Contentment", "Harmony", "Radiance", "Elegance",
                   "ArtisticBurst", "Like", "Good", "Nice", "Pleasant"]  # Moderate positive sentiments

negative_labels = ["Negative", "Anger", "Sadness", "Frustration", "Disappointment",
                   "Annoyance", "Resentment", "Bitterness", "Hate", "Bad", "Terrible",
                   "Awful", "Dislike", "Angry", "Upset"]  # Moderate negative sentiments

very_negative_labels = ["Frustrated", "Depressed","Gloomy", "Disgusted", "Anxious",
                        "Stressed", "Worried", "Scared", "Lonely", "Guilty", "Hurt",
                        "Embarrassed"]  # Strong negative sentiments

neutral_to_positive = ["Happy", "Great", "Amazing", "Excellent", "Wonderful", "Fantastic"]  # Neutral sentiments to move to Positive
neutral_to_negative = ["Bad", "Terrible", "Awful"]  # Neutral sentiments to move to Negative


# Standardize sentiment labels
df["Sentiment_Category"] = df["Sentiment"].astype(str).apply(
    lambda x: "Very Positive" if x.strip() in very_positive_labels else
              "Positive" if x.strip() in positive_labels or x.strip() in neutral_to_positive else
              "Negative" if x.strip() in negative_labels or x.strip() in neutral_to_negative else
              "Very Negative" if x.strip() in very_negative_labels else
              "Neutral"
)

# Remove empty sentiment values
df = df[df["Sentiment"].str.strip() != ""]

# Verify sentiment distribution
df["Sentiment_Category"].value_counts()
```

### Function to calculate polarity

```python
from textblob import TextBlob

# Function to calculate polarity using TextBlob
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

# Create a new 'Polarity' column
df['Polarity'] = df['Text'].apply(get_polarity)

# Function to categorize polarity into 5 sentiment categories
def categorize_polarity(polarity):
    if polarity >= 0.5:
        return "Very Positive"
    elif polarity > 0:
        return "Positive"
    elif polarity == 0:
        return "Neutral"
    elif polarity > -0.5:
        return "Negative"
    else:
        return "Very Negative"
```

###Visualisation: Top 10 hashtags with highest engagement (likes/retweets)


*   Purpose:

1. Identifies which hashtags generate the most engagement.
*   Insights:

1. Some hashtags drive significantly higher engagement than others.
2. Brands can leverage these hashtags for campaigns to maximize reach.

```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from collections import defaultdict
import numpy as np # Import the numpy library

# Create a dictionary to store hashtag engagement
hashtag_engagement = defaultdict(int)

# Iterate through the 'Hashtags' column
for index, row in df.iterrows():
    hashtags = row['Hashtags']
    if pd.notna(hashtags):  # Check if the cell is not NaN
        # Split hashtags while ensuring empty strings are removed
        individual_hashtags = [tag.strip() for tag in re.split(r'[ ,]+', hashtags) if tag.strip()]
        for hashtag in individual_hashtags:
            # Increment engagement count for each hashtag
            hashtag_engagement[hashtag] += row['Likes'] + row['Retweets']

# Convert the dictionary to a DataFrame for easier plotting
hashtag_engagement_df = pd.DataFrame.from_dict(hashtag_engagement, orient='index', columns=['Total Engagement'])

# Sort by total engagement and select top 10
top_hashtags = hashtag_engagement_df.nlargest(10, 'Total Engagement')

# Create the bar plot with the 'coolwarm' colormap
plt.figure(figsize=(12, 6))
plt.bar(top_hashtags.index, top_hashtags['Total Engagement'], color=plt.cm.coolwarm(np.linspace(0, 1, len(top_hashtags)))) # Now np is defined and can be used
plt.title('Top 10 Hashtags by Engagement (Unique Hashtags)')
plt.xlabel('Hashtag')
plt.ylabel('Total Engagement (Likes + Retweets)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

## Step 9: Virality & Trend Prediction

### Visualisation: Trend growth over time for top hashtags

*   Purpose:

1. Tracks which hashtags are gaining traction over time.

*   Insights:

1. A rising trend means the hashtag is gaining popularity and might go viral.

```python
# Explode hashtags: Convert comma/space-separated hashtags into individual rows
df["Hashtags"] = df["Hashtags"].fillna("")  # Fill NaN with empty string
df["Hashtags"] = df["Hashtags"].apply(lambda x: [tag.strip() for tag in re.split(r'[ ,]+', x) if tag.strip()])
df_exploded = df.explode("Hashtags")  # Create separate rows for each hashtag

# Group by date and hashtag, then count occurrences
hashtag_trend = df_exploded.groupby(["Date", "Hashtags"]).size().unstack().fillna(0)

# Select top 5 trending hashtags based on total mentions
top_trending_hashtags = hashtag_trend.sum().nlargest(5).index

# Plot trending hashtags over time
plt.figure(figsize=(15, 7))
hashtag_trend[top_trending_hashtags].plot(kind="line", marker="o", figsize=(12, 6))

plt.title("Virality: Top 5 Hashtags Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Mentions")
plt.legend(title="Hashtags", bbox_to_anchor=(1, 1), loc="upper left")
plt.xticks(rotation=45)
plt.show()
```

### Visualisation: Hourly/Daily engagement activity

*   Purpose:

1. Identifies best times for posting based on engagement activity.

*   Insights:

1. The brightest areas indicate the best times to post for maximum engagement.

```python
import seaborn as sns

# Extract hour and weekday
df["Hour"] = df["Timestamp"].dt.hour
df["Weekday"] = df["Timestamp"].dt.day_name()

# Create pivot table for heatmap
heatmap_data = df.pivot_table(index="Weekday", columns="Hour", values="Likes", aggfunc="sum")

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="coolwarm", annot=True, fmt=".0f")
plt.title("Engagement Activity Heatmap")
plt.xlabel("Hour of the Day")
plt.ylabel("Day of the Week")
plt.show()
```

## Visualisation: Top Hashtags Per Country would show:

Purpose:
* Identify the most frequently used hashtags in different regions.

Insights:

* Larger bubbles represent more frequently used hashtags.

* Helps businesses target regional campaigns using the most engaging hashtags.

```python
import plotly.express as px

# Preprocess hashtags: Split multiple hashtags and count occurrences per country
df["Hashtags"] = df["Hashtags"].fillna("").astype(str).str.findall(r"#\w+")

# Explode hashtags and count occurrences per country
hashtag_counts = df.explode("Hashtags").groupby(["Country", "Hashtags"])["Hashtags"].count().reset_index(name="Count")

# Get top 3 hashtags per country
top_hashtags = hashtag_counts.sort_values(["Country", "Count"], ascending=[True, False]).groupby("Country").head(3)


# Creating bubble chart
fig = px.scatter(top_hashtags,
                 x="Country",
                 y="Hashtags",
                 size="Count",
                 color="Hashtags",
                 title="Top Hashtags per Country",
                 labels={"Count": "Frequency", "Hashtags": "Top Hashtags"},
                 hover_data=["Count"],
                 template="plotly_white",  # Use a clean template
                 size_max=60,  # Adjust bubble size
                 color_discrete_sequence=px.colors.qualitative.Pastel)  # Use pastel colors

# Making it better for readability
fig.update_layout(
    xaxis_title="Country",
    yaxis_title="Hashtags",
    font=dict(family="Arial", size=12),
    legend_title_text="Hashtags",
    plot_bgcolor='rgba(0,0,0,0)'  # Transparent background
)

fig.show()
```
